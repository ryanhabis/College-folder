```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import bokeh

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```python
file_path = '/workspaces/College-folder/semester-1/Programming for Data Analytics/dataset/phoenix_park_weather.csv'
```

```python
weather_df = pd.read_csv(file_path)
weather_df.head()
```

```python
weather_df['date'] = pd.to_datetime(weather_df['date'], format='%d-%b-%y')
weather_df.set_index('date', inplace=True)

# Sorting the whole data
weather_df.sort_index(inplace=True)
weather_df.head()
```

```python
from bokeh.io import output_file
output_notebook()
output_file("max_temp.html")
```

```python
max_temp_plot = figure(title="Daily Maximum Temperature in Phonenix Park", 
                       x_axis_label='date', 
                       y_axis_label='Max temp in centigrade')

```

```python
max_temp_plot.line(x=weather_df.index[0:59], 
                   y= weather_df['2011-02':'2011-03'].maxt,
                   legend_label='max temp',
                   line_width=2,
                   color='red')
```

```python
show(max_temp_plot)
```

```python
from bokeh.plotting import figure, show

# Create a single plot for both max and min temperatures
temp_plot = figure(title="Daily Min and Max Temperatures in Phoenix Park",
                   x_axis_label='date',
                   y_axis_label='Temperature (C)')

# Min temperature line
temp_plot.line(x=weather_df.index[0:59],
               y=weather_df['2011-02':'2011-03'].mint,
               legend_label='Min Temp',
               line_width=2,
               color='blue')

# Max temperature line
temp_plot.line(x=weather_df.index[0:59],
               y=weather_df['2011-02':'2011-03'].maxt,
               legend_label='Max Temp',
               line_width=2,
               color='red')

# Show the combined plot
show(temp_plot)
```

```python
from bokeh.plotting import figure, show

new_plot = figure(
    title="Daily Min and Max Temperatures in Phoenix Park",
    x_axis_type='datetime',
    x_axis_label='date',
    y_axis_label='Temperature (C)'
)

x_vals = weather_df.index[0:59]
min_vals = weather_df.loc['2011-02':'2011-03'].mint
max_vals = weather_df.loc['2011-02':'2011-03'].maxt

new_plot.line(x=x_vals, y=min_vals, color='blue', line_width=1.5, legend_label='Min Temp')
new_plot.circle(x=x_vals, y=min_vals, color='blue', size=8, legend_label='Min Temp')

new_plot.line(x=x_vals, y=max_vals, color='red',  line_width=1.5, legend_label='Max Temp')
new_plot.circle(x=x_vals, y=max_vals, color='red',  size=8, legend_label='Max Temp')

show(new_plot)
```

```python
from bokeh.models import ColumnDataSource
output_file = 'new_plot.html'
source = ColumnDataSource(weather_df)

# Create a bar chart with weather_df
p = figure(x_axis_type='datetime')
p.circle(x='date', y='maxt', source=source, size=10, color='blue')

p.title.text = 'Maximum temperature v minimum temperature at Phoenix Park'
p.xaxis.axis_label = 'Minimum temperature (C)'
p.yaxis.axis_label = 'Maximum temperature (C)'
show(p)
```

```python
from bokeh.models import ColumnDataSource

output_file = 'new_plot.html'
source = ColumnDataSource(weather_df)

# Create a bar chart with weather_df
p = figure(x_axis_type='datetime')
p.circle(x='mint', y='maxt', source=source, size=10, color='blue')
p.line(x = 'mint', y = 'maxt', source=source, color='red')
p.title.text = 'Maximum temperature v minimum temperature at Phoenix Park'
p.xaxis.axis_label = 'Minimum temperature (C)'
p.yaxis.axis_label = 'Maximum temperature (C)'
show(p)
```

```python
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show, output_file
from bokeh.layouts import gridplot

output_file('new_plot.html')
source = ColumnDataSource(weather_df)

# Left chart
left = figure(title='Chart 1: Max vs Min Temperature')
left.line(x='date', y='maxt', source=source, color='red')
left.xaxis.axis_label = 'Minimum temperature (C)'
left.yaxis.axis_label = 'Maximum temperature (C)'

# Right chart  
right = figure(title='Chart 2: Max vs Min Temperature')
right.circle(x='date', y='maxt', source=source, size=10, color='blue')
right.xaxis.axis_label = 'Minimum temperature (C)'
right.yaxis.axis_label = 'Maximum temperature (C)'

p = gridplot([[left, right]])
show(p)
```

```python
from bokeh.models.tools import HoverTool

hover = HoverTool()

hover.tooltips=[
    ('Date', '@date'),
    ('Minimum temp', '@mint'),
    ('Max Temp', '@maxt'),
    ('Rainfall', '@rain')
]
p.add_tools(hover)
show(p)
```
