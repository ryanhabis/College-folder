```python
fig, ax = plt.subplots(2, 1, figsize=(12, 15))

sns.boxplot(y=weather['maxt'], x=weather['month'],data=weather ,palette='Set3', ax=ax[0])
sns.scatterplot(y=weather['mint'], x=weather['month'],data=weather ,palette='Set3',ax=ax[1])

plt.tight_layout()
plt.show()
```

```python
output_file("max_min_temp.html")
```

```python
temp_plot = figure(tools="pan,box_zoom,reset,save",
title="February and March 2011 max and min temperature in Phoenix Park",
x_axis_label='Day number', y_axis_label='Temperature (C)')
```

```python
temp_plot.line(x = list(range(59)), y = weather['2011-02':'2011-03'].maxt,
legend_label="Max temp", line_width = 2.5)
temp_plot.line(x = list(range(59)), y = weather['2011-02':'2011-03'].mint,
legend_label="Min temp", line_width = 2.5, line_color = 'red')
```

```python
temp_plot.circle(x = list(range(59)), y = weather['2011-02':'2011-03'].maxt,
legend_label="Max temp", fill_color = 'white', size = 6)
temp_plot.circle(x = list(range(59)), y = weather['2011-02':'2011-03'].mint,
legend_label="Min temp", fill_color = 'red', size = 6)
```

```python
show(temp_plot)
```
