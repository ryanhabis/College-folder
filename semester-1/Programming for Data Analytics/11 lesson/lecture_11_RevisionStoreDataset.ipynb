# Seaborn Data Visualization with Superstore Dataset
## Lecture 11 Revision

This notebook demonstrates various Seaborn visualization techniques using a sample of the Superstore dataset.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
%matplotlib inline
```

```python
# Load and sample data
file_path = "/home/ryanhabis/Documents/College folder/semester-1/Programming for Data Analytics/dataset/Dataset- Superstore (2015-2018).csv"
superstore = pd.read_csv(file_path)

# Convert dates
superstore['Order Date'] = pd.to_datetime(superstore['Order Date'])
superstore['Ship Date'] = pd.to_datetime(superstore['Ship Date'])

# Sample 500 rows for faster processing
sample_data = superstore.sample(n=500, random_state=42)

print(f"Dataset shape: {superstore.shape}")
print(f"Sample shape: {sample_data.shape}")
sample_data.head()
```

## 1. Scatterplot with Regression Line

```python
# Sales vs Profit with regression line
sns.lmplot(x='Sales', y='Profit', data=sample_data)
plt.xlabel('Sales ($)')
plt.ylabel('Profit ($)')
plt.title('Sales vs Profit Relationship')
plt.show()
```

## 2. Grouping with Hue Parameter

```python
# Color by Category
sns.lmplot(x='Sales', y='Profit', data=sample_data, hue='Category')
plt.xlabel('Sales ($)')
plt.ylabel('Profit ($)')
plt.title('Sales vs Profit by Category')
plt.show()
```

## 3. Subplots by Region

```python
# Separate plots for each region
sns.lmplot(x='Sales', y='Profit', data=sample_data, col='Region', col_wrap=2)
plt.suptitle('Sales vs Profit by Region', y=1.02)
plt.show()
```

## 4. Joint Plot

```python
# Scatterplot with histograms
sns.jointplot(x='Sales', y='Profit', data=sample_data, kind='scatter')
plt.suptitle('Joint Distribution: Sales vs Profit', y=1.02)
plt.show()
```

## 5. Boxplot and Violin Plot

```python
# Boxplot: Profit by Category
sns.boxplot(x='Category', y='Profit', data=sample_data)
plt.xlabel('Category')
plt.ylabel('Profit ($)')
plt.title('Profit Distribution by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Violin plot: Sales by Category
sns.violinplot(x='Category', y='Sales', data=sample_data)
plt.xlabel('Category')
plt.ylabel('Sales ($)')
plt.title('Sales Distribution by Category')
plt.xticks(rotation=45)
plt.show()
```

## 6. Correlation Analysis

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
corr_matrix = sample_data[['Sales', 'Quantity', 'Discount', 'Profit']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
# Create a 1x3 subplot grid for comprehensive comparison of sales data visualizations
# figsize=(18, 6) sets width to 18 inches and height to 6 inches for better visibility
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# STRIP PLOT (Left subplot) - displays individual data points
# alpha=0.6 adds transparency to see overlapping points, size=3 controls marker size
sns.stripplot(y='Sales', data=sample_data, ax=axes[0], alpha=0.6, size=3)
axes[0].set_title('Strip Plot: Individual Sales\n(Shows every data point)')
axes[0].set_ylabel('Sales Amount ($)')

# BOX PLOT (Middle subplot) - displays statistical summary with quartiles and outliers
# Shows median (line), interquartile range (box), and outliers (points)
sns.boxplot(y='Sales', data=sample_data, ax=axes[1])
axes[1].set_title('Box Plot: Sales Statistics\n(Median, Quartiles, Outliers)')
axes[1].set_ylabel('Sales Amount ($)')

# VIOLIN PLOT (Right subplot) - shows the probability density distribution
# Wider areas indicate where more data points are concentrated
sns.violinplot(y='Sales', data=sample_data, ax=axes[2])
axes[2].set_title('Violin Plot: Sales Distribution Shape\n(Probability Density)')
axes[2].set_ylabel('Sales Amount ($)')

# Adjust spacing between subplots to prevent overlap
plt.tight_layout()
# Display all three plots
plt.show()
```

```python
# Method 1: Using subplots (recommended)
fig, ax = plt.subplots(3, 1, figsize=(12, 15))

sns.stripplot(x='Category', y='Quantity', data=sample_data, palette="Set1", ax=ax[0])
sns.boxplot(x='Category', y='Quantity', data=sample_data, palette="Set3", ax=ax[1])
sns.violinplot(x='Category', y='Quantity', data=sample_data, palette="Set3", ax=ax[2])

plt.tight_layout()
plt.show()
```
