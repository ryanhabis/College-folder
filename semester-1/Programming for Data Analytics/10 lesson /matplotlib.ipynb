```python
import matplotlib.pyplot as plt

# Create a figure with 2 subplots (vertical arrangement)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 20))

# First chart (your original)
ax1.plot(imdb.votes, imdb.title, marker='1', linestyle='-', color='r')
ax1.set_xlabel("Votes")
ax1.set_ylabel("Movies")
ax1.set_title("Movies vs Votes")

# Second chart (example - you can customize this)
# Let's say you want to show rating vs title
ax2.plot(imdb.rating, imdb.title, marker='o', linestyle='--', color='b')
ax2.set_xlabel("Rating")
ax2.set_ylabel("Movies")
ax2.set_title("Movies vs Rating")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# First figure
plt.figure(figsize=(10, 20))
plt.plot(imdb.votes, imdb.title, marker='1', linestyle='-', color='r')
plt.xlabel("Votes")
plt.ylabel("Movies")
plt.title("Movies vs Votes")
plt.show()

# Second figure
plt.figure(figsize=(10, 20))
# Your second chart here - for example:
plt.scatter(imdb.rating, imdb.title, color='g', alpha=0.6)
plt.xlabel("Rating")
plt.ylabel("Movies")
plt.title("Movies vs Rating")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create subplots - your line chart and a bar chart
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 30))

# First chart (your original line plot)
ax1.plot(imdb.votes, imdb.title, marker='1', linestyle='-', color='r')
ax1.set_xlabel("Votes")
ax1.set_ylabel("Movies")
ax1.set_title("Movies vs Votes - Line Chart")

# Second chart - Horizontal bar chart (votes by movie)
ax2.barh(imdb.title, imdb.votes, color='skyblue')
ax2.set_xlabel("Votes")
ax2.set_ylabel("Movies")
ax2.set_title("Movies vs Votes - Bar Chart")
ax2.tick_params(axis='y', labelsize=8)  # Adjust font size if titles are long

plt.tight_layout()
plt.show()
```

```python
grouped_runtime = imdb.groupby("director")["runtime"].mean() 
# director is index, we can use it in x-ax

# Plot bar chart
```

```python
plt.figure(figsize=(30, 5))


plt.bar(grouped_runtime.index, grouped_runtime.values ,color = 'purple')

# Add labels
plt.xlabel("Director")
plt.ylabel("Average Runtime (minutes)")
plt.title("Average Film Runtime by Director")
plt.xticks(rotation=45) # Rotate x-axis labels if needed
```
