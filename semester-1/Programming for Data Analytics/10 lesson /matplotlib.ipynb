```python
# Import the matplotlib library and use the common alias 'plt' for easier reference
import matplotlib.pyplot as plt

# Create a figure with a 2x2 grid of subplots (4 total plots)
# fig: the overall figure container
# axes: a 2D array containing the 4 individual plot axes objects
# figsize=(15, 10): sets the overall figure size to 15 inches wide by 10 inches tall
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# --- FIRST ROW, FIRST COLUMN (Top-left plot) ---
# Create a boxplot of movie ratings in the first subplot (position [0,0])
axes[0,0].boxplot(imdb['rating'])
# Set the title for this specific subplot
axes[0,0].set_title('Rating Distribution')
# Set the y-axis label for this subplot
axes[0,0].set_ylabel('Rating')

# --- FIRST ROW, SECOND COLUMN (Top-right plot) ---
# Create a boxplot of movie runtimes in the second subplot (position [0,1])
axes[0,1].boxplot(imdb['runtime'])
# Set the title for this specific subplot
axes[0,1].set_title('Runtime Distribution')
# Set the y-axis label for this subplot
axes[0,1].set_ylabel('Minutes')

# --- SECOND ROW, FIRST COLUMN (Bottom-left plot) ---
# Calculate the frequency of each genre and get the top 5 most common genres
# value_counts(): counts how many times each unique genre appears
# head(5): returns only the top 5 most frequent genres
genre_counts = imdb['genre'].value_counts().head(5)
# Create a bar chart in the third subplot (position [1,0])
# genre_counts.index: the genre names (x-axis labels)
# genre_counts.values: the count values (y-axis heights)
axes[1,0].bar(genre_counts.index, genre_counts.values)
# Set the title for this specific subplot
axes[1,0].set_title('Top 5 Genres')
# Rotate the x-axis labels 45 degrees to prevent overlapping with long genre names
axes[1,0].tick_params(axis='x', rotation=45)

# --- SECOND ROW, SECOND COLUMN (Bottom-right plot) ---
# Calculate the frequency of each director and get the top 5 most common directors
# value_counts(): counts how many times each unique director appears
# head(5): returns only the top 5 most frequent directors
director_counts = imdb['director'].value_counts().head(5)
# Create a bar chart in the fourth subplot (position [1,1])
# director_counts.index: the director names (x-axis labels)
# director_counts.values: the count values (y-axis heights)
axes[1,1].bar(director_counts.index, director_counts.values)
# Set the title for this specific subplot
axes[1,1].set_title('Top 5 Directors')
# Rotate the x-axis labels 45 degrees to prevent overlapping with long director names
axes[1,1].tick_params(axis='x', rotation=45)

# Automatically adjust the spacing between subplots to prevent overlapping elements
# This ensures proper padding between plots, labels, and titles
plt.tight_layout()

# Display the complete figure with all 4 subplots
plt.show()
```
