## Bar Charts

- Bar charts show information in a similar way to the line plots, but with bars instead of points.
- Use plt.bar(x,y) to plot a bar chart
- They are used mainly to display one variable
- Specify the variable whose height you want to be represented by bars on the y axis (second argument of plt.bar).
- The x axis is usually the number index.

```python
imdb.head()
```

```python
help(plt.bar)
```

```python
plt.bar(imdb.index, imdb.runtime)
plt.show()
```

There is a bit too much information here.

For illustration purposes, I will focus on the top 10 movies to show how barplots can be formatted.

```python
plt.bar(imdb.index[0:10], imdb.runtime[0:10])
plt.show()
```

Change the width of the bars using the width=

```python
plt.bar(imdb.index[0:10], imdb.runtime[0:10], width = 0.5)
plt.show()
```

Change the alignment of the bars in relation to their x coordinate using the align

- 'center': Center the base on the x positions (default value).
- 'edge': Align the left edges of the bars with the x positions.

```python
plt.bar(imdb.index[0:10], imdb.runtime[0:10], width = 0.5, align = 'edge')
plt.show()
```

Specify the value at which the bars begin on the y axis using the bottom=

The y coordinate(s) of the bottom side(s) of the bars.

```python
plt.bar(imdb.index[0:10], imdb.runtime[0:10], 
        width = 0.5, align = 'edge', bottom = 100)
plt.show()
```

- Again, we can change the color of the bars using the color argument.
- Note that all of the arguments for these functions are specified on their help pages. I am not fully formatting every plot to be the finished article.
- How could the plot below be improved?

```python
plt.bar(imdb.index[0:10], imdb.runtime[0:10],
        width = 0.5, bottom = 90, color = 'purple')
plt.show()
```

```python
# Group by director and compute the average runtime

grouped_runtime = imdb[0:5].groupby("director")["runtime"].mean() # director is index, we can use it in x-axis.

# Plot bar chart
plt.bar(grouped_runtime.index, grouped_runtime.values)

# Add Labels
plt.xlabel("Director")
plt.ylabel("Average Runtime (minutes)")
plt.title("Average Film Runtime by Director")
plt.xticks(rotation=45) # Rotate x-axis Labels if needed

plt.show()
```

## Boxplots

- Boxplots are used to show the distribution of one or more variables.
- Boxplots are often used to detect outliers as they are shown clearly on the pt.
- Boxplots also display the median and upper and lower quartiles of the data.
- The command for a boxplot is plt.boxplot(values).

```python
imdb.head()
```

```python
plt.boxplot(imdb.rating)
plt.show()
```

```python
plt.boxplot(imdb.metascore)
plt.show()
```

- I got an empty plot when using metascore column from the imdb DataFrame as a boxplot.
- This is due to missing data.

```python
sum(pd.isnull(imdb.metascore))
```

Use pd.notnull as a "mask" to only plot the non-NAN values from metascore column

```python
mask = pd.notnull(imdb.metascore)
imdb.metascore[mask]
```

```python
plt.boxplot(imdb.metascore[mask])
plt.show()
```

Draw the boxplot horizontally using the `vert = value False`

```python
plt.boxplot(imdb.metascore[mask], vert = False)
plt.show()
```

Boxplots can be used to show two different continuous variables on the same plot.

However, if the variables are not on similar scales, it may not be easy to compare their distributions:

```python
plt.boxplot([imdb.rating, imdb.metascore[mask]])
plt.show()
```

Multiply rating by 10 to put them onto the same scale (out of 100).

```python
rating_100 = imdb.rating * 10
plt.boxplot([rating_100, imdb.metascore[mask]])
plt.show()
```

We now see that metascore column has a lower median and a wider distribution than rating.

Label the boxplots using plt.xticks . The first argument is a list of where the ticks appear. The second argument is a list of the labels.

```python
plt.boxplot([rating_100, imdb.metascore[mask]])
plt.xticks([1,2], ['Rating', 'Metascore'])
plt.show()
```

Boxplots for splitting a continuous variable by the levels of a categorical variable are not straightforward using matplotlib. They are easier in seaborn, which we will see next week.

## Creating multiple plots in one window

fig, ax = plt.subplots() creates a figure with an array of axis objects that can be worked on.

```python
fig, ax = plt.subplots()
plt.show()
```

Create a figure and one subplot by specifying no argument in plt.subplots.

Use `ax.plot` to add a line plot to the figure. We are editing the axis object. We will see that certain functions are performed on the axis object (ax), while others are performed on the figure object (fig):

- fig: A reference to the Figure object, which is the "container" for your plots.
- ax: A reference to the Axes object, which represents a single subplot (the area where data is plotted).

```python
fig, ax = plt.subplots()
ax.plot(imdb.rating)
plt.show()
```

Add a second line to the plot

```python
fig, ax = plt.subplots()
ax.plot(imdb.rating)
ax.plot(imdb.metascore)
plt.show()
```

Plot the two lines on the same scale

```python
fig, ax = plt.subplots()
ax.plot(rating_100)
ax.plot(imdb.metascore)
plt.show()
```

```python
# Create a figure with a 2x2 grid of subplots (4 plots total)
# fig: the overall figure container that holds everything
# axis: a 2D numpy array containing the 4 individual plot axes objects
# The axis array will be indexed as [row, column]
fig, axis = plt.subplots(2, 2)

# --- FIRST ROW, FIRST COLUMN (Top-left plot) ---
# Create a scatter plot using the plot() function with no connecting lines
# rating_100: the data for x-axis (likely ratings scaled to 100)
# marker='o': use circle markers for each data point
# linestyle='None': don't connect the points with lines (makes it a scatter plot)
# color='r': set the marker color to red
axis[0, 0].plot(rating_100, marker='o', linestyle='None', color='r')
# Set the y-axis label for this subplot
axis[0, 0].set_ylabel("rating")

# --- FIRST ROW, SECOND COLUMN (Top-right plot) ---
# Create another scatter plot using the plot() function
# imdb.metascore: the data for x-axis (Metacritic scores)
# marker='v': use triangle-down markers for each data point
# linestyle='None': don't connect points with lines
# color='b': set the marker color to blue
axis[0, 1].plot(imdb.metascore, marker='v', linestyle='None', color='b')
# Set the y-axis label for this subplot
axis[0, 1].set_ylabel("metascore")

# --- SECOND ROW, FIRST COLUMN (Bottom-left plot) ---
# Create a proper scatter plot comparing two variables
# rating_100: x-axis data (ratings scaled to 100)
# imdb.metascore: y-axis data (Metacritic scores)
# This shows the relationship between user ratings and critic scores
axis[1, 0].scatter(rating_100, imdb.metascore)

# --- SECOND ROW, SECOND COLUMN (Bottom-right plot) ---
# Create a line plot of rating_100 data with some transparency
# rating_100: the data to plot
# alpha=0.7: set transparency to 70% opaque (30% transparent)
axis[1, 1].plot(rating_100, alpha=0.7)
# Add a second line plot on the same subplot for metascore data
# imdb.metascore: the second dataset to plot
# alpha=0.7: same transparency level
# Both lines will be plotted on the same axes for comparison
axis[1, 1].plot(imdb.metascore, alpha=0.7)

# Add a main title for the entire figure (all 4 subplots)
# fig.suptitle: creates a title above all the subplots
# fontsize=12: sets the title font size
fig.suptitle("Example of different plots", fontsize=12)

# Automatically adjust the spacing between subplots and around the edges
# Prevents overlapping of labels, titles, and plot elements
# tight_layout() automatically calculates optimal spacing
fig.tight_layout()

# Display the complete figure with all 4 subplots
plt.show()
```
